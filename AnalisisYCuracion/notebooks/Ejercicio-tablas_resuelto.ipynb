```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
tabla_1 = pd.read_csv("../input/wtageinf.csv")
tabla_1.head(10)
```

**Gráficos**

A continuación se muestran los gráficos de crecimiento separados por sexo.

```python
tabla_1[tabla_1.Sex==1].plot(x='Agemos',
                             y=['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'],
                             title='Tabla de crecimiento para niños',
                             grid=True,
                             figsize=(10,6))
```

```python
tabla_1[tabla_1.Sex==2].plot(x='Agemos',
                             y=['P3', 'P5', 'P10', 'P25', 'P50', 'P75','P90', 'P95', 'P97'],
                             title='Tabla de crecimiento para niñas',
                             grid=True,
                             figsize=(10,6))
```
